- Using .loc   <== label index

```python
colours.iloc[1] # <== positionl ke-2
```

```python
colours.iloc[0] # <== positional ke-1
```

```python
# Cari tim dari LA yang bermain tandang di tahun 92

df[
    (df['team_id'].str.startswith("LA")) &
    (df['year_id'] == 1992) &
    (df['notes'].notnull())
    
]
```

```python
# cari tim dari dallas yang bermain di kandang dan menang dari tahun 1995 - 2012 with score > 100

dallasmaverics = df[
    (df['team_id'] == 'DAL') &
    (df['fran_id'] == 'Mavericks') &
    (df['game_location'] == "A") &
    (df['game_result'] == 'W') &
    (df['pts'] > 100) &
    ((df['year_id'] >= 1995) & (df['year_id'] <= 2012))
    
    
    
]

dallasmaverics
```

### Grouping and Aggregation

```python
df
```

```python
df['pts'].sum()
```

```python
df['pts'].max()
```

```python
df['pts'].mean()
```

```python
#cari tau total point dari masing-masing tim

df.groupby('fran_id', sort=False)['pts'].sum() # <== default dari sort adalah 'True' dan  akan melakukan pengurutan berdasarkan abjad
```

```python
df.groupby(['fran_id', 'team_id'])['pts'].sum() # <== default dari sort adalah 'True' dan  akan melakukan pengurutan berdasarkan abjad
```

```python
# Golden State Warrior tahun 2014/2015. berapa jumlah kemenangan dan kekalahan di tahun tersebut

df[
    (df['fran_id'] == 'Warriors') &
    (df['year_id'] == 2015)
    
    
].groupby(['year_id', 'game_result'])['game_id'].count()
```

```python
# berapa jumlah kemenangan dan kekalahan tim Lakers mulai tahun 1940

df[
    (df['fran_id'] == 'Lakers') &
    (df['year_id'] >= 1940)
    
    
].groupby(['year_id', 'team_id', 'fran_id', 'game_result'])['game_id'].count()
```

### Manipulasi Kolom

tujuannya :
1. cleansing data awal
2. untuk mencapai insight yang diinginkan

bisa dengan cara add dan atau drop kolom

```python
nba = df.copy()
```

```python
nba.shape
```

```python
#add kolom baru 'selisih_poin

nba['selisih_poin'] = nba['pts'] - nba['opp_pts']
nba.head(10)
```

```python
#merubah nama kolom game_result --> result, game_location --> venue

ubah_kolom = nba.rename(columns={'game_result': "result", 'game_location': 'venue'})

ubah_kolom.head(10)
```

```python
nba.shape
```

```python
nba.drop(['selisih_poin'], inplace=True, axis=1)
nba.shape
```

```python
nba.head(10)
```

```python
nba.axes
```
