## Case-3. compare top 5 immigrate to canada

```python
#sort dulu by Total
pd_canada_sort = pd_canada.sort_values(by='Total', ascending=False,axis=0)

#ambil 5 row data dulu saja
Top5toCanada = pd_canada_sort.head(5)
Top5toCanada


```

```python
#ambil kolom years saja
Top5toCanada_Years = Top5toCanada[years]
Top5toCanada_Years
```

```python
#transpose datanya
Top5toCanada_Years.transpose()
```

```python
#plot
Top5toCanada_Years.transpose().plot();
```

```python
#coba rapikan
Top5toCanada_Years.transpose().plot(figsize=(20,10));

plt.title('Top 5 Country migrate to Canada')
plt.xlabel('Years')
plt.ylabel('no. immigrants')


plt.show();
```

# tipe plot
- line
- area
- histogram
- bar ==> vertical or horizontal
- pie
- Box
- Scatter
- Bubble

```python
#area plot pada top 5 country
Top5toCanada_Years.transpose().plot(figsize=(20,10), kind='area', stacked=False, alpha=0.25); # <== alpha adalah transparan value

plt.title('Top 5 Country migrate to Canada')
plt.xlabel('Years')
plt.ylabel('no. immigrants')


plt.show();
```

```python
#5 area tersedikit yang masuk ke Canada
Top5LowtoCanada = pd_canada.tail(5)
Top5LowtoCanada
```

```python
Top5LowtoCanada_Years = Top5LowtoCanada[years].transpose()
```

```python
Top5LowtoCanada_Years
```

```python
Top5LowtoCanada_Years.plot(figsize=(20,10),kind='area',alpha=0.45,stacked=True);
plt.title('Lowest 5 Country to Canada')
plt.xlabel('Years')
plt.ylabel('no immigrants')


plt.show()
```

### Histogram

```python
#distribusi frekuensi yang masuk ke Canada pada tahun 2013 terhadap bermacam negara
pd_canada[2013].head()
```

```python
count, bin_edges = np.histogram(pd_canada[2013])

print(count)
print(bin_edges)
```

```python
pd_canada[2013].plot(kind='hist',figsize=(15,8));

plt.title('histogram immigrant from 2013 from various country')
plt.xlabel('no. immigrant')
plt.ylabel('no. country')

plt.show();
```

```python
pd_canada[2013].plot(kind='hist',figsize=(15,8), xticks = bin_edges);

plt.title('histogram immigrant from 2013 from various country')
plt.xlabel('no. immigrant')
plt.ylabel('no. country')

plt.show();
```

## PIE Chart

```python
#contoh per continent
df_continent = pd_canada.groupby('Continent', axis=0).sum()
df_continent
```

```python
df_continent['Total'].plot(kind='pie', startangle=60, autopct="%1.1f%%");
plt.axis('equal')

plt.show();
```

```python
df_continent['Total'].plot(
    kind="pie",
    figsize=(15,8),
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink'],
    explode=[0.1,0,0,0,0.1,0.1],
    pctdistance=1.12,
    labels=None
);


plt.title("Total immigrants to Canada 1980-2013 by Continent", y=1.12)

# memastikan size graph nya sesuai figsize
plt.axis('equal')

# tambahin legent menggunakan script
plt.legend(labels=df_continent.index, loc="upper left")

plt.show()
```

# subplot or FIGURE
memasukkan beberapa plot ke dalam satu gambar (figure) yang sama

```python
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1) # <== 1 row, 2 kolom, plot ke-1
ax1 = fig.add_subplot(1,2,2) # <== 1 row, 2 kolom, plot ke-2


#plot ke-1
China_India.transpose().plot(kind='box', figsize=(18,7), color='blue', vert=False, ax=ax0)
ax0.set_title('Immigrant China India to Canada from 1980-2013')
ax0.set_ylabel('country')
ax0.set_xlabel('no. immigrants')

#plot ke-2
China_India.transpose().plot(kind='line', figsize=(18,7), ax=ax1)
ax1.set_title('Immigrant China India to Canada from 1980-2013')
ax1.set_xlabel('years')
ax1.set_ylabel('no. immigrants')

plt.show();
```

## Scatter plot

```python
NDS = pd_canada.loc[['Denmark','Sweden','Norway'],years].transpose()
NDS_Total = pd.DataFrame(NDS.sum(axis=1))
NDS_Total.reset_index(inplace=True) # kita ubah kolom index kita menjadi kolom biasa
NDS_Total.columns=['year','total']
NDS_Total['year']=NDS_Total['year'].astype(int)
NDS_Total.head()
```

```python
NDS_Total.plot(kind='scatter', x='year', y='total', figsize=(10,7), color='darkblue');

plt.title('Norway, Denmark, Sweden Immigrant to Canada')
plt.xlabel('years')
plt.ylabel('no. immigrants')


plt.show();
```
